# Task 1: Conceptual Understanding 
### 1. Difference between "Love" and "love" in NLP
- "Love" and "love" are treated as different tokens if case normalization is not applied.
- NLP models may interpret them as separate words.
- Converting text to lowercase ensures consistency and better frequency analysis.

### 2. What happens if stopwords are not removed?
- Increases noise in the dataset
- Reduces model performance
- Important words may get less importance
- Increases computation unnecessarily
  
### 3. When removing stopwords is harmful?
- Scenario 1: Sentiment Analysis
  - Sentence: "I am not happy"
  - Removing "not" changes meaning →  Wrong sentiment
- Scenario 2: Question Answering
  - Sentence: "What is your name?"
  - Removing "what" breaks meaning
  
### 4. Stemming vs Lemmatization
Feature	       Stemming	                 Lemmatization\
Method	    Cuts words	                Uses vocabulary\
Output	    May be invalid	            Always meaningful\
Example	   "running"→"run"/"runn"	    "running"→"run"

# Task 2:Build Advanced Preprocessing Function

```python
import re

def preprocess_text(text):
    if not text or not isinstance(text, str):
        return [], ""

    # Lowercase
    text = text.lower()

    # Remove URLs & emails
    text = re.sub(r"http\S+|www\S+|@\S+", "", text)

    # Remove numbers
    text = re.sub(r"\d+", "", text)

    # Handle repeated characters (soooo → so)
    text = re.sub(r"(.)\1{2,}", r"\1", text)

    # Remove special characters & emojis
    text = re.sub(r"[^\w\s]", "", text)

    # Remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()

    # Tokenization
    tokens = text.split()

    # Remove short tokens (<=2) except "no", "not"
    cleaned_tokens = [
        word for word in tokens 
        if len(word) > 2 or word in ["no", "not"]
    ]

    cleaned_sentence = " ".join(cleaned_tokens)

    return cleaned_tokens, cleaned_sentence
```

# Task 3: Stress Testing


```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

results = []

for sentence in test_sentences:
    tokens, cleaned = preprocess_text(sentence)
    print("\nOriginal:", sentence)
    print("Tokens:", tokens)
    print("Cleaned:", cleaned)
    
    results.append((sentence, tokens, cleaned))
```

# Task 4: Token Analytics


```python
def token_analytics(tokens):
    if not tokens:
        return 0, 0, 0
    
    total = len(tokens)
    unique = len(set(tokens))
    avg_length = sum(len(word) for word in tokens) / total
    
    return total, unique, round(avg_length, 2)


for sentence, tokens, cleaned in results:
    total, unique, avg = token_analytics(tokens)
    
    print("\nSentence:", sentence)
    print("Total Tokens:", total)
    print("Unique Tokens:", unique)
    print("Avg Length:", avg)
```

# Task 5: Frequency Analysis

```python
from collections import Counter

all_tokens = []

for _, tokens, _ in results:
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("Top 10 Most Frequent Words:")
print(counter.most_common(10))

print("\nTop 5 Least Frequent Words:")
print(counter.most_common()[-5:])
```

# Task 6: Build Full Pipeline

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, cleaned = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(cleaned)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }


output = full_pipeline(test_sentences)

print("\nPipeline Output:")
print(output)
```

# Task 7: Error Handling

```python
edge_cases = [
    "",
    "😂😂😂😂",
    "1234567890"
]

for case in edge_cases:
    tokens, cleaned = preprocess_text(case)
    print("\nInput:", case)
    print("Tokens:", tokens)
    print("Cleaned:", cleaned)
```
